```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.isaaa.org/animalbiotechdatabase/ngtlist/ngt/default.asp?NGTID=2&RegProcess=0"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

```

```python
table = soup.find('table')

rows = table.find_all('tr')

data = []

for row in rows:
    cells = row.find_all('td')
    if len(cells) < 5:
        continue
    
    country = cells[1].get_text(strip=True)
    
    # Animal: separate common and scientific names
    animal_common = cells[2].contents[0].strip() if cells[2].contents else ""
    animal_scientific = cells[2].em.get_text(strip=True) if cells[2].em else ""
    
    gene = cells[3].get_text(strip=True)
    year = cells[0].get_text(strip=True)  # first cell is year
    details_url = "https://www.isaaa.org" + cells[4].a['href'] if cells[4].a else ""
    
    data.append({
        "year": year,
        "country": country,
        "animal_common": animal_common,
        "animal_scientific": animal_scientific,
        "gene": gene,
        "details_url": details_url
    })

df = pd.DataFrame(data)
print("Total rows scraped:", len(df))
df.head()

df.to_csv("isaa_animalbiotech_full.csv", index=False)
print("CSV saved as 'isaa_animalbiotech_full.csv'")
```

```python
import time

df['Institute'] = ""
df['Editing_Method'] = ""
df['Delivery'] = ""
df['Foreign_DNA'] = ""
df['Regulatory_Process'] = ""

for idx, url in enumerate(df.loc[df['details_url'] != "", 'details_url']):
    print(f"Scraping details page {idx+1}/{len(df)}: {url}")
    
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    
    def get_field_value(field_name):
        td_label = soup.find('td', text=lambda x: x and field_name in x)
        if td_label:
            td_value = td_label.find_next_sibling('td')
            if td_value:
                return td_value.get_text(strip=True)
        return ""
    
    df.loc[idx, 'Institute'] = get_field_value('Institute')
    df.loc[idx, 'Editing_Method'] = get_field_value('Editing Method')
    df.loc[idx, 'Delivery'] = get_field_value('Delivery')
    df.loc[idx, 'Foreign_DNA'] = get_field_value('Foreign DNA')
    df.loc[idx, 'Regulatory_Process'] = get_field_value('Undergone Regulatory Process')
    
    time.sleep(0.5)

df.head()
```

```python
df.to_csv("isaa_animalbiotech_full_details.csv", index=False)

print("Saved full dataset with details to 'isaa_animalbiotech_full_details.csv'")
print("Total rows:", len(df))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.isaaa.org/animalbiotechdatabase/ngtlist/ngt/default.asp?NGTID=2&RegProcess=1"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the main data table
table = soup.find('table')

# Extract all rows
rows = table.find_all('tr')

data = []

for row in rows:
    cells = row.find_all('td')
    if len(cells) < 5:
        continue
    
    year = cells[0].get_text(strip=True)
    country = cells[1].get_text(strip=True)
    
    # Split common and scientific names
    animal_common = cells[2].contents[0].strip() if cells[2].contents else ""
    animal_scientific = cells[2].em.get_text(strip=True) if cells[2].em else ""
    
    gene = cells[3].get_text(strip=True)
    details_url = "https://www.isaaa.org" + cells[4].a['href'] if cells[4].a else ""
    
    data.append({
        "year": year,
        "country": country,
        "animal_common": animal_common,
        "animal_scientific": animal_scientific,
        "gene": gene,
        "details_url": details_url
    })

df2 = pd.DataFrame(data)
print("Basic table scraped — rows found:", len(df2))
df2.head()
```

```python
df2['Institute'] = ""
df2['Editing_Method'] = ""
df2['Delivery'] = ""
df2['Foreign_DNA'] = ""
df2['Regulatory_Process'] = ""
```

```python
import time

def get_field_value(soup, field_name):
    td_label = soup.find('td', text=lambda x: x and field_name in x)
    if td_label:
        td_val = td_label.find_next_sibling('td')
        if td_val:
            return td_val.get_text(strip=True)
    return ""

for idx, url in enumerate(df2.loc[df2['details_url'] != "", 'details_url']):
    print(f"({idx+1}/{len(df2)}) Fetching {url}")
    resp = requests.get(url)
    detail_soup = BeautifulSoup(resp.text, 'html.parser')
    
    df2.loc[idx, 'Institute'] = get_field_value(detail_soup, 'Institute')
    df2.loc[idx, 'Editing_Method'] = get_field_value(detail_soup, 'Editing Method')
    df2.loc[idx, 'Delivery'] = get_field_value(detail_soup, 'Delivery')
    df2.loc[idx, 'Foreign_DNA'] = get_field_value(detail_soup, 'Foreign DNA')
    df2.loc[idx, 'Regulatory_Process'] = get_field_value(detail_soup, 'Undergone Regulatory Process')
    
    time.sleep(0.4)  # polite delay

df2.head()
```

```python
df2.to_csv("isaa_animalbiotech_regprocess1_full.csv", index=False)
print("Saved as 'isaa_animalbiotech_regprocess1_full.csv' (rows:", len(df2), ")")
```
